```python
import pandas as pd

df = pd.read_csv('bank-full.csv', sep=';')
df.head()
```

```python
total = len(df)
customers = df[df['y'] == 'yes'].shape[0]

print("Total:", total)
print("Customers:", customers)
print("Conversion Rate:", round((customers/total)*100,2), "%")
```

```python
pd.crosstab(df['contact'], df['y'], normalize='index') * 100
```

```python
pd.crosstab(df['month'], df['y'], normalize='index') * 100
```

```python
pd.crosstab(df['job'], df['y'], normalize='index') * 100
```

```python
import matplotlib.pyplot as plt

df['y'].value_counts().plot(kind='bar')
plt.title('Conversion Results')
plt.show()
```

**Insights**

-Conversion rate is low with high drop-off

-Cellular channel performs better

-Some months outperform others

-Certain job groups convert more
